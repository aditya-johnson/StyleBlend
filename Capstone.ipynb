```python
from flask import Flask, request, jsonify, render_template_string, send_from_directory
from pyngrok import ngrok
import threading
import os
import shutil

# Initialize Flask app
app = Flask(__name__)

# Paths for storing images
celebrity_path = "/content/drive/My Drive/Style-Your-Hair/Celebrity_imgs/source.png"
ffhq_path = "/content/drive/My Drive/Style-Your-Hair/ffhq_image/source.png"
target_path = "/content/drive/My Drive/Style-Your-Hair/ffhq_image"
output_directory = "/content/drive/My Drive/Style-Your-Hair/style_your_hair_output"
similar_images_folder = "/content/drive/My Drive/Style-Your-Hair/sims"
trending_hairstyles_folder = "/content/drive/My Drive/Style-Your-Hair/trending_hairstyles"
target_image_path = os.path.join(target_path, "target.png")

# Serve static images for display in the app
@app.route('/get_similar_images')
def get_similar_images():
    try:
        # List all PNG files in both directories
        similar_images = [f"/sims/{img}" for img in os.listdir(similar_images_folder) if img.endswith(".png")]
        trending_images = [f"/trending/{img}" for img in os.listdir(trending_hairstyles_folder) if img.endswith(".png")]

        # Return as JSON
        return jsonify({
            "similar_images": similar_images,
            "trending_images": trending_images
        })
    except Exception as e:
        return jsonify({"error": str(e)}), 500



# HTML template with updated design, countdown, and show images button
html_template = """
<!doctype html>
<html lang="en">
<head>
    <title>Upload Image - Futuristic Design</title>
    <style>
        body {
            display: flex;
            justify-content: center;
            align-items: center;
            flex-direction: column;
            background-color: #141414;
            color: #c7c7c7;
            font-family: Arial, sans-serif;
            height: 100vh;
            margin: 0;
        }
        h2 {
            color: #00ffc3;
            text-shadow: 0 0 8px #00ffc3, 0 0 12px #00ffc3;
        }
        .container {
            background: #1f1f1f;
            border: 1px solid #333;
            border-radius: 8px;
            padding: 20px;
            box-shadow: 0 0 15px rgba(0, 255, 195, 0.2);
            width: 300px;
            text-align: center;
        }
        input[type="file"] {
            display: none;
        }
        label {
            display: inline-block;
            padding: 10px 20px;
            color: #00ffc3;
            border: 2px solid #00ffc3;
            border-radius: 4px;
            cursor: pointer;
            text-shadow: 0 0 8px #00ffc3;
        }
        button {
            background: #00ffc3;
            color: #141414;
            border: none;
            padding: 10px 20px;
            border-radius: 4px;
            cursor: pointer;
            margin-top: 20px;
            box-shadow: 0 0 8px #00ffc3;
        }
        button:hover {
            background-color: #00d4a3;
        }
        #fileName {
            margin-top: 10px;
            color: #00ffc3;
        }
        .countdown {
            margin-top: 20px;
        }
        #timer {
            font-size: 1.2em;
            margin-top: 10px;
            color: #ff4d4d;
            text-shadow: 0 0 8px #ff4d4d;
        }
        .images-container {
            display: flex;
            flex-wrap: wrap;
            gap: 10px;
            justify-content: center;
            margin-top: 20px;
        }
        .images-container img {
            width: 80px;
            height: 80px;
            cursor: pointer;
            border: 2px solid #333;
            border-radius: 4px;
        }
        .images-container img.selected {
            border-color: #00ffc3;
        }
    </style>
    <script>
        let initialTimeLeft = 60; // 5 minutes for initial countdown
        let finalTimeLeft = 1800; // 30 minutes for final countdown
        let selectedImage = "";

        let initialCountdownInterval, finalCountdownInterval;

        function startInitialCountdown() {
            initialCountdownInterval = setInterval(function() {
                initialTimeLeft--;
                let minutes = Math.floor(initialTimeLeft / 60);
                let seconds = initialTimeLeft % 60;
                document.getElementById("initialTimer").innerHTML = minutes + ":" + (seconds < 10 ? "0" + seconds : seconds);

                if (initialTimeLeft <= 0) {
                    clearInterval(initialCountdownInterval);
                    fetchSimilarImages();
                }
            }, 1000);
        }

        function startFinalCountdown() {
            finalCountdownInterval = setInterval(function() {
                finalTimeLeft--;
                let minutes = Math.floor(finalTimeLeft / 60);
                let seconds = finalTimeLeft % 60;
                document.getElementById("finalTimer").innerHTML = minutes + ":" + (seconds < 10 ? "0" + seconds : seconds);

                if (finalTimeLeft <= 0) {
                    clearInterval(finalCountdownInterval);
                }
            }, 1000);
        }

       function fetchSimilarImages() {
        fetch("/get_similar_images")
            .then(response => response.json())
            .then(data => {
                // Similar Images Section
                const similarImagesContainer = document.getElementById("similarImagesContainer");
                similarImagesContainer.innerHTML = ""; // Clear old images
                data.similar_images.forEach(src => {
                    const img = document.createElement("img");
                    img.src = src;
                    img.alt = "Similar Image";
                    img.onclick = () => selectImage(src);
                    similarImagesContainer.appendChild(img);
                });

                // Trending Hairstyles Section
                const trendingImagesContainer = document.getElementById("trendingImagesContainer");
                trendingImagesContainer.innerHTML = ""; // Clear old images
                data.trending_images.forEach(src => {
                    const img = document.createElement("img");
                    img.src = src;
                    img.alt = "Trending Image";
                    img.onclick = () => selectImage(src);
                    trendingImagesContainer.appendChild(img);
                });

                // Display both sections
                document.getElementById("similarImagesSection").style.display = "block";
                document.getElementById("trendingImagesSection").style.display = "block";
            })
            .catch(error => alert("Error fetching images: " + error));
    }

        function selectImage(src) {
            selectedImage = src;

            // Highlight the selected image
            const images = document.querySelectorAll(".images-container img");
            images.forEach(img => img.classList.remove("selected"));
            event.target.classList.add("selected");

            // Show the confirm button after selection
            document.getElementById("confirmSelectionSection").style.display = "block";
        }

        function uploadImage() {
            const formData = new FormData();
            const imageFile = document.getElementById("image").files[0];

            if (!imageFile) {
                alert("Please select a file to upload.");
                return;
            }

            formData.append("image", imageFile);

            fetch("/upload", {
                method: "POST",
                body: formData
            })
            .then(response => {
                if (!response.ok) {
                    return response.json().then(data => {
                        throw new Error(data.message);
                    });
                }
                return response.json();
            })
            .then(data => {
                alert(data.message);
                if (data.message === "Image saved successfully in both locations") {
                    document.getElementById("initialCountdownSection").style.display = "block";
                    startInitialCountdown();
                }
            })
            .catch(error => alert("Error: " + error.message));
        }

        function confirmSelection() {
            if (!selectedImage) {
                alert("Please select an image.");
                return;
            }
            fetch("/select_image", {
                method: "POST",
                headers: { "Content-Type": "application/json" },
                body: JSON.stringify({ image_path: selectedImage })
            })
            .then(response => response.json())
            .then(data => {
                if (data.message === "Image selection successful") {
                    document.getElementById("finalCountdownSection").style.display = "block";
                    startFinalCountdown();
                    document.getElementById("showImagesBtn").style.display = "inline-block";
                } else {
                    alert("Failed to select image: " + data.error);
                }
            })
            .catch(error => alert("Error: " + error));
        }
    </script>
</head>
<body>
    <div class="container">
        <h2>Upload an Image</h2>
        <label for="image">Choose File</label>
        <input type="file" id="image" name="image" required onchange="document.getElementById('fileName').innerText = 'Selected file: ' + this.files[0].name;">
        <p id="fileName">No file chosen</p>
        <button onclick="uploadImage()">Upload Image</button>

        <div id="initialCountdownSection" class="countdown" style="display: none;">
            <h3>Initial 2-Minute Countdown</h3>
            <p id="initialTimer">02:00</p>
        </div>

        <div id="similarImagesSection" style="display: none;">
            <h3>Select a Similar Image</h3>
            <div class="images-container" id="similarImagesContainer"></div>
        </div>

        <div id="trendingImagesSection" style="display: none;">
            <h3>Select a Trending Hairstyle</h3>
            <div class="images-container" id="trendingImagesContainer"></div>
        </div>

        <div id="confirmSelectionSection" style="margin-top: 20px; display: none;">
            <button onclick="confirmSelection()">Confirm Selection</button>
        </div>

        <div id="finalCountdownSection" class="countdown" style="display: none;">
            <h3>Final 30-Minute Countdown</h3>
            <p id="finalTimer">30:00</p>
        </div>

        <button id="showImagesBtn" onclick="window.location.href='/show_images'" style="display: none;">
            Show Images
        </button>
    </div>
</body>
</html>

"""

@app.route('/')
def home():
    return render_template_string(html_template)

@app.route('/upload', methods=['POST'])
def upload_image():
    if 'image' not in request.files:
        return jsonify({"error": "No image file found"}), 400

    image_file = request.files['image']

    if image_file.filename == '':
        return jsonify({"error": "No selected file"}), 400

    try:
        image_data = image_file.read()
        with open(celebrity_path, 'wb') as f:
            f.write(image_data)
        with open(ffhq_path, 'wb') as f:
            f.write(image_data)

        return jsonify({"message": "Image saved successfully in both locations"}), 200
    except Exception as e:
        return jsonify({"error": str(e)}), 500




@app.route('/sims/<filename>')
def serve_similar_images(filename):
    return send_from_directory(similar_images_folder, filename)

@app.route('/trending/<filename>')
def serve_trending_images(filename):
    return send_from_directory(trending_hairstyles_folder, filename)



@app.route('/select_image', methods=['POST'])
def select_image():
    image_path = request.json.get("image_path")
    if not image_path:
        return jsonify({"error": "Image path not provided"}), 400

    try:
        # Define the full path of the selected image
        selected_image_full_path = os.path.join(similar_images_folder, os.path.basename(image_path))

        # Copy the selected image to `target_path` as `target.png`
        shutil.copy(selected_image_full_path, target_image_path)

        return jsonify({"message": "Image selection successful"}), 200
    except Exception as e:
        return jsonify({"error": str(e)}), 500

@app.route('/show_images')
def show_images():
    # List all PNG images in the output directory
    png_images = [f for f in os.listdir(output_directory) if f.endswith('.png')]
    if png_images:
        # Render HTML to display each image
        images_html = ''.join([f'<img src="/image/{image}" width="200px" style="margin:10px; border-radius:8px; box-shadow:0 0 8px #00ffc3;">' for image in png_images])
        return f"<h2 style='color:#00ffc3; text-align:center;'>Generated Images:</h2><div style='text-align:center;'>{images_html}</div>"
    else:
        return "<h2 style='color:#ff4d4d; text-align:center;'>No images found</h2>"
@app.route('/image/<filename>')
def serve_image(filename):
    return send_from_directory(output_directory, filename)

ngrok_url = None

def run_flask():
    global ngrok_url
    public_url = ngrok.connect(5003)  # Change to 5002 to match Flask app port
    ngrok_url = public_url.public_url  # Store the ngrok URL in the global variable
    print("Ngrok tunnel URL:", ngrok_url)  # Display the public URL
    app.run(port=5003)

# Start the Flask app in a separate thread
flask_thread = threading.Thread(target=run_flask)
flask_thread.start()
```

```python
import os
import shutil
import numpy as np
from deepface import DeepFace
from scipy.spatial.distance import cosine, euclidean, cityblock
from IPython.display import display, Image
import torch
import torchvision.transforms as T
from torchvision import models
from PIL import Image as PILImage
from torch import nn

# Paths
MODEL_PATH = "/content/drive/MyDrive/Style-Your-Hair/best_model (1).pth"
user_image_path = '/content/drive/My Drive/Style-Your-Hair/Celebrity_imgs/source.png'
base_folder_path = '/content/drive/My Drive/Style-Your-Hair/Celebrity_imgs'
save_folder = '/content/drive/My Drive/Style-Your-Hair/sims'

# Class mappings for face shapes
class_to_idx = {
    "Oval": 0,
    "Round": 1,
    "Square": 2,
    "Heart": 3,
    "Long": 4
}
idx_to_class = {v: k for k, v in class_to_idx.items()}

# Load and setup the face shape classifier
device = "cuda" if torch.cuda.is_available() else "cpu"
model = models.efficientnet_b4(weights=None)
num_classes = len(class_to_idx)
model.classifier = nn.Sequential(
    nn.Dropout(p=0.3, inplace=True),
    nn.Linear(model.classifier[1].in_features, num_classes)
)
model.load_state_dict(torch.load(MODEL_PATH, map_location=device))
model = model.to(device)
model.eval()

# Transformation for input images
inference_transform = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

def classify_face_shape(image_path):
    """
    Classifies the face shape of the given image.
    """
    try:
        img = PILImage.open(image_path).convert("RGB")
        input_tensor = inference_transform(img).unsqueeze(0).to(device)
        with torch.no_grad():
            output = model(input_tensor)
            _, predicted_idx = torch.max(output, 1)
            return idx_to_class[predicted_idx.item()]
    except Exception as e:
        print(f"Error classifying image {image_path}: {e}")
        return None

def get_face_embedding(image_path):
    """
    Extracts face embedding from the given image path using DeepFace.
    """
    try:
        analysis = DeepFace.represent(image_path, model_name='VGG-Face', enforce_detection=False)
        return analysis[0]['embedding']
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")
        return None

def calculate_composite_similarity(embedding1, embedding2, weights=(0.4, 0.4, 0.2)):
    """
    Calculates composite similarity score between two face embeddings.
    """
    if embedding1 is None or embedding2 is None:
        return 0

    cos_sim = 1 - cosine(embedding1, embedding2)
    euclid_dist = euclidean(embedding1, embedding2)
    manhattan_dist = cityblock(embedding1, embedding2)

    # Normalize distances and calculate weighted composite score
    norm_euclid = 1 / (1 + euclid_dist)
    norm_manhattan = 1 / (1 + manhattan_dist)
    composite_score = (weights[0] * cos_sim) + (weights[1] * norm_euclid) + (weights[2] * norm_manhattan)
    return composite_score

def find_most_similar_in_shape(user_image_path, shape_folder_path):
    """
    Finds the top 5 most similar celebrities to the user image within the specified shape folder.
    """
    user_embedding = get_face_embedding(user_image_path)
    if user_embedding is None:
        raise ValueError(f"Could not get embedding for user image: {user_image_path}")

    similarities = []
    for celeb_image in os.listdir(shape_folder_path):
        celeb_image_path = os.path.join(shape_folder_path, celeb_image)

        try:
            celeb_embedding = get_face_embedding(celeb_image_path)
            similarity = calculate_composite_similarity(user_embedding, celeb_embedding)
            similarities.append((celeb_image_path, similarity))
        except Exception as e:
            print(f"Error processing {celeb_image_path}: {e}")

    # Sort by similarity and return top 5 results
    return sorted(similarities, key=lambda item: item[1], reverse=True)[:5]

# Step 1: Classify the face shape of the user image
predicted_shape = classify_face_shape(user_image_path)
if predicted_shape:
    print(f"Predicted face shape: {predicted_shape}")

    # Step 2: Find top 5 similar images within the predicted face shape folder
    shape_folder_path = os.path.join(base_folder_path, predicted_shape)
    top_5_similar = find_most_similar_in_shape(user_image_path, shape_folder_path)

    # Save the top 5 images with similarity scores and face shape in the filename
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    for idx, (celeb_path, score) in enumerate(top_5_similar, start=1):
        celeb_name = os.path.basename(celeb_path)
        save_name = f"{predicted_shape}_{score:.4f}_{celeb_name}"
        save_path = os.path.join(save_folder, save_name)
        shutil.copy(celeb_path, save_path)
        print(f"Saved: {save_name}")
else:
    print("Failed to classify face shape.")

```
